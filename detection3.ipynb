```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
image=cv2.imread('OCR.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap='gray')
path="C:/Users/fateme/Desktop/PyClass/PyCourse-29/ocr_1"
```

```python
# thresh=255-image
thresh=cv2.adaptiveThreshold(image,100,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,5,4)
plt.imshow(thresh,cmap='gray')
```

```python
contours,_=cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
name=100
for c in contours:
    x,y,w,h=cv2.boundingRect(c)
    name+=1
    temp=image[y:y+h,x:x+w]
    cv2.imwrite(os.path.join(path,f"{name}.jpg"),temp )
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,0),2)
plt.imshow(image,cmap='gray')
```
