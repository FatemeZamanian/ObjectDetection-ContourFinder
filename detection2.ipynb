```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread('gerdalies.tif',cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap='gray')
```

```python
# thresh=cv2.adaptiveThreshold(image,100,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,9,8)
thresh=255-image
plt.imshow(thresh,cmap='gray')
```

```python
contours,_=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
for c in contours:
    white_area=cv2.contourArea(c)
    if white_area<1000:
        x,y,w,h=cv2.boundingRect(c)
        image[y:y+h,x:x+w]=255
plt.imshow(image,cmap='gray')     
```
